```python
yaml_content = """# Ultralytics YOLO 🚀, AGPL-3.0 license
# YOLO11 object detection model with P2-P5 outputs. For Usage examples see https://docs.ultralytics.com/tasks/detect
# from: https://github.com/ultralytics/ultralytics/pull/16558/files

# Parameters
nc: 1 # number of classes
scales: # model compound scaling constants, i.e. 'model=yolo11n.yaml' will call yolo11.yaml with scale 'n'
  # [depth, width, max_channels]
  s: [0.50, 0.50, 1024]
  m: [0.50, 1.00, 512]
  l: [1.00, 1.00, 512]
  x: [1.00, 1.50, 512]

# YOLO11-P2 backbone
backbone:
  # [from, repeats, module, args]
  - [-1, 1, Conv, [64, 3, 2]] # 0-P1/2
  - [-1, 1, Conv, [128, 3, 2]] # 1-P2/4
  - [-1, 2, C3k2, [256, False, 0.25]]
  - [-1, 1, Conv, [256, 3, 2]] # 3-P3/8
  - [-1, 2, C3k2, [512, False, 0.25]]
  - [-1, 1, Conv, [512, 3, 2]] # 5-P4/16
  - [-1, 2, C3k2, [512, True]]
  - [-1, 1, Conv, [1024, 3, 2]] # 7-P5/32
  - [-1, 2, C3k2, [1024, True]]
  - [-1, 1, SPPF, [1024, 5]] # 9
  - [-1, 2, C2PSA, [1024]] # 10

# YOLO11-P2 head
head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 6], 1, Concat, [1]] # cat backbone P4
  - [-1, 2, C3k2, [512, False]] # 13

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 4], 1, Concat, [1]] # cat backbone P3
  - [-1, 2, C3k2, [256, False]] # 16 (P3/8-small)

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 2], 1, Concat, [1]] # cat backbone P2
  - [-1, 2, C3k2, [128, False]] # 19 (P2/4-xsmall)

  - [-1, 1, Conv, [128, 3, 2]]
  - [[-1, 16], 1, Concat, [1]] # cat head P4
  - [-1, 2, C3k2, [256, False]] # 22 (P3/8-small)

  - [-1, 1, Conv, [256, 3, 2]]
  - [[-1, 13], 1, Concat, [1]] # cat head P4
  - [-1, 2, C3k2, [512, False]] # 25 (P4/16-medium)

  - [-1, 1, Conv, [512, 3, 2]]
  - [[-1, 10], 1, Concat, [1]] # cat head P5
  - [-1, 2, C3k2, [1024, True]] # 28 (P5/32-large)

  - [[19, 22, 25, 28], 1, Detect, [nc]] # Detect(P3, P4, P5)
"""

# Ghi đè file dataset.yaml
with open("/kaggle/working/yolov11-p2.yaml", "w") as f:
    f.write(yaml_content)
```

```python
from ultralytics import YOLO

# ==============================================================================
# CHIẾN LƯỢC: SPEED RUN + AGGRESSIVE AUGMENTATION
# ==============================================================================

# Load model P2 (tốt cho vật nhỏ)
model = YOLO('/kaggle/working/yolov11-p2.yaml').load('yolo11s.pt')

results = model.train(
    data='/kaggle/working/yolo_dataset/dataset.yaml',
    
    # --- CẤU HÌNH TỐC ĐỘ ---
    epochs=100,
    imgsz=640,
    batch=32,
    cache=True,
    workers=8,
    device=0,
    single_cls=True,
    
    # --- CẤU HÌNH LOSS ---
    box=8.0,
    cls=0.2,
    dfl=1.5,
    
    # --- CẤU HÌNH AUGMENTATION (THEO YÊU CẦU CỦA BẠN) ---
    # Màu sắc: Biến thiên mạnh để chống lệch domain giữa các video
    hsv_h=0.015,
    hsv_s=0.7,    # Tăng cường thay đổi độ đậm nhạt màu
    hsv_v=0.4,    # Tăng cường thay đổi độ sáng
    
    # Hình học
    degrees=15,   # Xoay nhẹ (Drone hay bị nghiêng)
    translate=0.2,
    scale=0.6,    # [ĐIỀU CHỈNH]: Giảm từ 0.9 xuống 0.6. 
                  # Lý do: scale 0.9 sẽ làm mất vật thể nhỏ (điện thoại) khi zoom out.
    shear=5.0,
    perspective=0.0,
    
    # Lật ảnh (Rất tốt cho Drone view)
    flipud=0.5,   # Lật dọc (Ok vì góc nhìn từ trên cao)
    fliplr=0.5,   # Lật ngang
    
    # Ghép ảnh
    mosaic=1.0,
    mixup=0.15,
    copy_paste=0.3,
    
    # --- CHIẾN THUẬT TRAINING ---
    close_mosaic=20,    # 10 epoch cuối tắt augmentation để chốt hạ mAP
    patience=100,
    
    # Optimizer
    optimizer='AdamW',
    lr0=0.001,                
    cos_lr=True,
    
    name='yolov11_p2_aggressive_aug'
)
```
