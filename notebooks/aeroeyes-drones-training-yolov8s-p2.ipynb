```python
yaml_content = """train: /kaggle/working/yolo_dataset/images/train
val: /kaggle/working/yolo_dataset/images/val
nc: 1
names:
- target_object
"""

# Ghi Ä‘Ã¨ file dataset.yaml
with open("/kaggle/working/yolo_dataset/dataset.yaml", "w") as f:
    f.write(yaml_content)
```

```python
from ultralytics import YOLO

model = YOLO('/kaggle/working/yolov8-p2.yaml').load('yolov8s.pt')
```

```python
results = model.train(
    data='/kaggle/working/yolo_dataset/dataset.yaml',
    
    # --- Cáº¤U HÃŒNH Tá»C Äá»˜ ---
    epochs=100,
    imgsz=640,
    batch=32,
    cache=True,
    workers=8,
    device=0,
    single_cls=True,
    
    # --- Cáº¤U HÃŒNH LOSS ---
    box=8.0,
    cls=0.2,
    dfl=1.5,
    
    # --- Cáº¤U HÃŒNH AUGMENTATION (THEO YÃŠU Cáº¦U Cá»¦A Báº N) ---
    # MÃ u sáº¯c: Biáº¿n thiÃªn máº¡nh Ä‘á»ƒ chá»‘ng lá»‡ch domain giá»¯a cÃ¡c video
    hsv_h=0.015,
    hsv_s=0.7,    # TÄƒng cÆ°á»ng thay Ä‘á»•i Ä‘á»™ Ä‘áº­m nháº¡t mÃ u
    hsv_v=0.4,    # TÄƒng cÆ°á»ng thay Ä‘á»•i Ä‘á»™ sÃ¡ng
    
    # HÃ¬nh há»c
    degrees=15,   # Xoay nháº¹ (Drone hay bá»‹ nghiÃªng)
    translate=0.2,
    scale=0.6,    # [ÄIá»€U CHá»ˆNH]: Giáº£m tá»« 0.9 xuá»‘ng 0.6. 
                  # LÃ½ do: scale 0.9 sáº½ lÃ m máº¥t váº­t thá»ƒ nhá» (Ä‘iá»‡n thoáº¡i) khi zoom out.
    shear=5.0,
    perspective=0.0,
    
    # Láº­t áº£nh (Ráº¥t tá»‘t cho Drone view)
    flipud=0.5,   # Láº­t dá»c (Ok vÃ¬ gÃ³c nhÃ¬n tá»« trÃªn cao)
    fliplr=0.5,   # Láº­t ngang
    
    # GhÃ©p áº£nh
    mosaic=1.0,
    mixup=0.15,
    copy_paste=0.3,
    
    # --- CHIáº¾N THUáº¬T TRAINING ---
    close_mosaic=20,    # 10 epoch cuá»‘i táº¯t augmentation Ä‘á»ƒ chá»‘t háº¡ mAP
    patience=100,
    
    # Optimizer
    optimizer='AdamW',
    lr0=0.001,                
    cos_lr=True,
    
    name='yolov11_p2_aggressive_aug'
)
```
